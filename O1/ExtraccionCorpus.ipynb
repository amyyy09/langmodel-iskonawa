```python
def map_morphemes_to_words(words_line, morphemes_line):
    # Split the lines into words and morphemes
    words = words_line.split()
    morphemes = morphemes_line.split()
    
    # Initialize an empty list to store the mapping
    mapping = []
    
    # Initialize indices for words and morphemes
    word_index = 0
    morpheme_index = 0
    
    # Iterate through the morphemes and map them to words
    while morpheme_index < len(morphemes):
        current_word = words[word_index]
        current_morpheme = morphemes[morpheme_index]
        
        # Add the current morpheme to the mapping
        mapping.append((current_word, current_morpheme))
        
        # Move to the next morpheme
        morpheme_index += 1
        
        # Check if the current word is fully mapped
        if morpheme_index < len(morphemes) and (morphemes[morpheme_index].startswith('=') or morphemes[morpheme_index].startswith('-')):
            continue
        else:
            # Move to the next word
            word_index += 1
    
    return mapping

# Example usage
words_line = "awen awinin yoa akai yoa ahon"
morphemes_line = "awen awin =n yoa ak -ai yoa ak -hon"

mapping = map_morphemes_to_words(words_line, morphemes_line)
for word, morpheme in mapping:
    print(f"Word: {word}, Morpheme: {morpheme}")
```
