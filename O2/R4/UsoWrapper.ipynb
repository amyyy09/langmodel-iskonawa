```python
map_list = []
for entry in multi_corpus.entries:
    words_mapped = entry.map_words_with_glosses()
    map_list.append(words_mapped)
```

```python
seen_words = set()

# Initialize an empty list to store the new list of dictionaries
new_list = []

# Iterate over the original list of dictionaries
for mapping in map_list:
    for word, gloss in mapping.items():
        if word not in seen_words:
            # Add the word to the seen set
            seen_words.add(word)
            # Add the word and its gloss to the new list as a dictionary
            new_list.append({word: gloss})

# Print the new list of dictionaries
for entry in new_list:
    print(entry)
```

```python
# Write the new list of dictionaries to a text file, without the curly braces
with open("word_gloss_pairs.txt", "w", encoding="utf-8") as file:
    for entry in new_list:
        for word, gloss in entry.items():
            file.write(f"{word}: {gloss}\n")
```
